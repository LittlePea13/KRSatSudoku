```python
paired_test = np.zeros((len(measures),len(heuristics),len(heuristics)))
mean_test = np.zeros((len(measures),len(heuristics),len(heuristics)))

for k, measure in enumerate(measures):
    for i, heur in enumerate(heuristics):
        for j, heur2 in enumerate(heuristics): 
            paired_test[k,i,j] = stats.wilcoxon(heur[measure], heur2[measure]).pvalue
            mean_test[k,i,j] = stats.mannwhitneyu(heur[measure], heur2[measure]).pvalue
```
